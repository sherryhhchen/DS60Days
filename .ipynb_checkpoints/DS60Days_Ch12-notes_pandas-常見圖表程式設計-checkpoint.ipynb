## 1.折線圖

```python
%matplotlib inline
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### 繪製單一折線圖

```python
ts = pd.Series(np.random.randn(200), index = pd.date_range('1/1/2020', periods =200))

#繪製單一折線圖
ts = ts.cumsum() #累計總和
print('以表格呈現數據：\n', ts)
ts.plot()
```

### 繪製多筆資料的折線圖

```python
#繪製多筆資料的折線圖 (多個欄位，每個欄位為一筆資料)

df = pd.DataFrame(np.random.randn(200, 3), index=pd.date_range('1/1/2020', periods=200), columns=['A','B','C'])
df = df.cumsum()
print(df)
df.plot()
```

## 圓餅圖

```python
#繪製圓餅圖
df = pd.Series(np.random.rand(4), index = ['A','B','C','D'], name = 'title')
print(df)
df.plot.pie()
```

## 長條圖

```python
#繪製一般長條圖
df = pd.DataFrame(np.random.rand(10,2), columns=['A', 'B'])

#印出資料集
print(df)

#繪製長條圖
df.plot.bar()
```

```python
#繪製堆疊長條圖
df = pd.DataFrame(np.random.rand(10,2), columns=['A', 'B'])

#印出資料集
print(df)

#繪製堆疊長條圖
df.plot.bar(stacked=True)
```

## 箱型圖

```python
#創建資料
df = pd.DataFrame(np.random.rand(10, 2), columns=['A', 'B'])

#印出資料集
print(df)

#繪製箱型圖
df.boxplot()
```

## 散佈圖

```python
#創建資料集
df = pd.DataFrame(np.random.rand(10, 2), columns = ['A', 'B'])

#印出資料集
print(df)

#繪製散佈圖
df.plot.scatter(x='A', y='B')
```
