## 1.折線圖

```python
%matplotlib inline
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### 繪製單一折線圖

```python
ts = pd.Series(np.random.randn(200), index = pd.date_range('1/1/2020', periods =200))

#繪製單一折線圖
ts = ts.cumsum() #累計總和
print('以表格呈現數據：\n', ts)
ts.plot()
```

### 繪製多筆資料的折線圖

```python
#繪製多筆資料的折線圖 (多個欄位，每個欄位為一筆資料)

df = pd.DataFrame(np.random.randn(200, 3), index=pd.date_range('1/1/2020', periods=200), columns=['A','B','C'])
df = df.cumsum()
print(df)
df.plot()
```
